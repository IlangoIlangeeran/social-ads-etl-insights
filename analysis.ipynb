```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the cleaned data
df = pd.read_csv("cleaned_ads.csv")

# Display the first few rows
df.head()
```

```python
sns.histplot(data=df, x="Age", hue="Purchased", multiple="stack", palette="Set2")
plt.title("Purchase Count by Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(x="Purchased", y="EstimatedSalary", data=df)
plt.title("Salary Distribution by Purchase")
plt.xlabel("Purchased")
plt.ylabel("Estimated Salary")
plt.show()
```

```python
sns.countplot(x="Purchased", data=df)
plt.title("Purchase Distribution")
plt.xticks([0, 1], ['Not Purchased', 'Purchased'])
plt.show()
```
